```python
pointsdf = pd.DataFrame(

{
    '1st_point': [(3,4), (-6,5), (0,0), (-10,1), (4,5)],
    '2nd_point': [(-3,4), (0,0), (2,2), (10,10), (1,1)]
}

)
```

```python
pointsdf
```

```python
def euclidean(dat):
    point1 = dat['1st_point']
    point2 = dat['2nd_point']
    return (((point2[0] - point1[0])**2) + ((point2[1] - point1[1])**2))**0.5
```

```python
pointsdf.apply(euclidean, axis = 1)
```
