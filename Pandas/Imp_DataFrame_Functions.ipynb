## sort values with null values

```python
students = {
    'name' : ['ishaq', 'pooja', 'suma', 'chaitu'],
    'marks' : [12,3,4,np.NAN]
}
```

```python
students
```

```python
tempdf = pd.DataFrame(students)
```

```python
tempdf
```

```python
tempdf.sort_values('marks')
```

```python
tempdf.sort_values('marks', ascending = False)
```

### if you see in the top that, regardless of ascending true/false, the null values are coming at the end, we can control this behaviour


```python
tempdf.sort_values('marks', na_position = 'first')
```

```python
tempdf.sort_values('marks', ascending = False, na_position = 'first')
```

```python
#inplace
tempdf.sort_values('marks', ascending = False, na_position = 'first', inplace=True)
```

```python
tempdf
```

```python
tempdf = pd.DataFrame(students)
```
