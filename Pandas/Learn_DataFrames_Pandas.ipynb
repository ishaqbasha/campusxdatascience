```python
data = {
 'A': [1, 2, 3],
 'B': [4, 5, 6],
 'C': [7, 8, 9]
}
df = pd.DataFrame(data)
print(df)
```

```python
value = df.iloc[1, 2] # Retrieves the value at the second row and third column
print(value)
```

```python
df = pd.DataFrame(data, index=['X', 'Y', 'Z'])
print(df)
```

```python
value = df.loc['Y', 'B']
```

```python
value
```

```python
'A': [1, 2, 3],
 'B': [4, 5, 6],
 'C': [7, 8, 9]
}
df = pd.DataFrame(data)

# Slicing rows and columns using iloc
subset = df.iloc[1:3, 0:2] # Retrieves rows 2 and 3, columns 1 and 2
print(subset)
```
