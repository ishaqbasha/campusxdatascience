```python
movies[movies['Series_Title'].str.startswith('A')].value_counts(subset='Genre')
```

```python
# find ranking of each movies in the group according to the IMDB Score

def movieRank(group):
    group['Genre_Rating'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genres.apply(movieRank)
```

```python
# Find the normalized rating of each movie with respect to the genre group

def normal(group):
    group['Normal_Rating']  =  (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normal)
```

```python
# Find top 10 batsmans who scored runs
batsmans = ipl.groupby('batsman')
```

```python
batsmans['batsman_runs'].sum().sort_values(ascending = False).head(10)
```

```python
# find the batsman with max no of sixes

sixes = pd.Series(dtype='int64')
for group, data in batsmans:
    sixes[group] = (data['batsman_runs'] == 6).sum()
```

```python
sixes.sort_values(ascending = False).head(1)
```

```python
# other way to solve the above problem

six = ipl[ipl['batsman_runs'] == 6]
six.value_counts('batsman')
```

```python
six.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

```python
six.groupby('batsman').size().sort_values(ascending=False)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
new_data = ipl[((ipl['batsman_runs'] == 6) | (ipl['batsman_runs'] == 4)) & (ipl['over'] > 15)]
```

```python
new_data.groupby('batsman')['batsman'].count().sort_values(ascending = False)
```

```python
# find virat kohli's record against all teams

vkMatches = ipl[ipl['batsman'] == 'V Kohli']
```

```python
vkMatches.head()
```

```python
# create a function that can return the highest score of any batsman using groupby

def highest(player):
    temp_batsman = ipl[ipl['batsman'] == player]
    return temp_batsman.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False)
```

```python
highest('V Kohli')
```
